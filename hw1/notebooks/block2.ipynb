```python
%%file task1.py
# %%file is an Ipython magic function that saves the code cell as a file

from mrjob.job import MRJob # import the mrjob library
from mrjob.step import MRStep

class MRTalkativeChars(MRJob):
    def steps(self):
        return [
            MRStep(mapper=self.mapper,
                   reducer=self.reducer_sum),
            MRStep(reducer=self.reducer_agg)
        ]        
    
    def mapper(self, _, line):
        cols = line.split()
        if len(cols) > 2:
            yield (cols[1].replace('"', ''), 1)
        
    def reducer_sum(self, char, values):
        yield None, (char, sum(values))

    def reducer_agg(self, _, values):
        top = {}
        for key, val in values:
            if key in top:
                top[key] += val
            else:
                top[key] = val
        for line in reversed(sorted(top.items(), key=lambda x:x[1])[-20:]):
            yield line[0], line[1]
        
if __name__ == "__main__":
    MRTalkativeChars.run()
```

```python
%%file task2.py
# %%file is an Ipython magic function that saves the code cell as a file

from mrjob.job import MRJob # import the mrjob library
from mrjob.step import MRStep

class MRMaxlen(MRJob):
    def steps(self):
        return [
            MRStep(mapper=self.mapper,
                   reducer=self.reducer_max),
            MRStep(reducer=self.reducer_final)
        ]        
        
    def mapper(self, _, line):
        lines = line.split("\n")
        char = {}
        for line in lines:
            cols = line.split('" "')
            if len(cols) > 2:

                if cols[1] in char:
                    char[cols[1]] = cols[2] if len(char[cols[1]].split()) > len(cols[2].split()) else char[cols[1]]
                else:
                    char[cols[1]] = cols[2]

        for key, val in char.items():
            yield (key, val)
    
    def reducer_max(self, word, values):
        max = ''
        for val in values:
            if len(val.split()) > len(max.split()):
                max = val
        yield None, (word, max)

    def reducer_final(self, _, values):
        for key, val in reversed(sorted(values, key=lambda x: len(x[1]))):
            yield key, val
        
if __name__ == "__main__":
    MRMaxlen.run()
```

```python
%%file task3.py
# %%file is an Ipython magic function that saves the code cell as a file
# import subprocess
# import sys
# subprocess.check_call([sys.executable, "-m", "pip", "install", 'nltk'])

from mrjob.job import MRJob # import the mrjob library
from mrjob.step import MRStep
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.util import bigrams
from collections import Counter
import string


class MRMaxlen(MRJob):
    def steps(self):
        return [
            MRStep(mapper_init=self.mapper_init,
                   mapper=self.mapper,
                   reducer=self.reducer_init),
            MRStep(reducer=self.reducer_final)
        ]        
    def mapper_init(self):
        nltk.download('punkt')
        nltk.download('stopwords')

    def mapper(self, _, line):

        lines = line.split("\n")
        bigram_list = []
        for line in lines:
            if len(line.split()) > 2:
                _, _, text = line.split('" "')
                
                # clean_text = "".join([i for i in text if i not in string.punctuation]).lower().strip().split()
                clean_text = "".join([i for i in text if i not in string.punctuation]).lower().strip()
                tokens = word_tokenize(clean_text)
                stop_words = set(stopwords.words('english'))
                filtered = []

                for token in tokens:
                    if token not in stop_words:
                        filtered += [token]

                bigram_list += bigrams(filtered)
                
        for bigram, cnt in Counter(bigram_list).items():
            yield (bigram, int(cnt))
    
    def reducer_init(self, word, values):
        yield None, (word, sum(values))

    def reducer_final(self, _, values):
        for val in reversed(sorted(values, key=lambda x:x[1])[-20:]):
            yield " ".join(val[0]), val[1]


if __name__ == "__main__":
    MRMaxlen.run()
```
